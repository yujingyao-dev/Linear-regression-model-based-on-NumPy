```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1.四张图画在一起，2x2

```python
# 定义自变量取值范围
x = np.linspace(-5, 5, 201)
y = np.linspace(-5, 5, 201)
# 生成网格
x, y = np.meshgrid(x, y)
print(x) # 纵向复制
print(y) # 横向复制
# 两个矩阵一叠加，就是每一点的坐标
```

```python
# 定义函数关系
f1 = np.sin(x) * np.cos(y)
f2 = (x**4 + y**4)**0.5
f3 = np.tanh(x) * np.tanh(y)
```

```python
# 创建格式
fig = plt.figure(figsize=(8,8)) # 2行2列 figure只获得一个fig对象，subplot还会有一个ax对象，一个2x2的矩阵

# 为了3D，必须额外写一个ax
ax1 = fig.add_subplot(2, 2, 1, projection='3d')
ax1.plot_surface(x, y, f1, cmap='vanimo')
ax1.set_xlabel("x") # 区别于plt.xlabel
ax1.set_ylabel("y")
ax1.set_zlabel("f(x,y)")
ax1.set_title("f(x,y)=sinx+cosy")
# ax1.set(zticklabels=[]) # 取消z轴线标签

ax2 = fig.add_subplot(2, 2, 2, projection='3d')
ax2.plot_surface(x, y, f2, cmap='RdYlBu')
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax1.set_zlabel("f(x,y)")
ax2.set_title("f(x,y)=(x^4+y^4)^0.5")

ax3 = fig.add_subplot(2, 2, 3, projection="3d")
ax3.plot_surface(x, y, f3, cmap='RdYlGn')
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.set_zlabel("f(x,y)")
ax3.set_title("f(x,y)=tanhx+tanhy")

# 绘制等高线图
ax4 = fig.add_subplot(2, 2, 4)
CS = ax4.contour(x, y, f3, levels=10)
ax4.clabel(CS, fontsize=8)
ax4.set_xlabel("x")
ax4.set_ylabel("y")
ax4.set_title("f(x,y)=tanhx+tanhy")
```
